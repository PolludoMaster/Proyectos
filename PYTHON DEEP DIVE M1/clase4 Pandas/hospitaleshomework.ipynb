<body>
<p>latitude</p>
<p>longitude</p>
<p>name</p>
<p>label</p>
</body>
<footer>Con los correspondientes datos extraídos del CSV original, el campo name tiene que corresponder con la dirección del hospital, y el campo label con el nombre del hospital.</footer>

```python
import pandas as pd
```

```python
hospitales = pd.read_csv('hospitales2.csv', sep = ',')
```

```python
hospitales[['LONGITUD','LATITUD']] = hospitales.WKT.str.split("-",expand=True)[[1,2]]
```

```python
hospitales['LATITUD']=hospitales['LATITUD'].str.replace(')','')
hospitales['LONGITUD']=hospitales['LONGITUD'].str.replace('(','')
hospitales['LONGITUD']=hospitales['LONGITUD'].str.replace(' ','')
```

```python
hospitales['LATITUD'] = ('-' + hospitales['LATITUD'])
hospitales['LONGITUD'] = ('-' + hospitales['LONGITUD'])
```

```python
hospitales['LATITUD'] = pd.to_numeric(hospitales['LATITUD'], downcast='float')
hospitales['LONGITUD'] = pd.to_numeric(hospitales['LONGITUD'], downcast='float')
```

```python
hospitales.head()
```

```python
hospitales2 = pd.DataFrame(hospitales,columns=['LONGITUD','LATITUD','NOMBRE','DOM_GEO','WEB'])
```

```python
hospitales2.to_csv(r'hospitales.csv', index = False)
```
